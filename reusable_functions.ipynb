```python
import pandas
```

```python
tall = pandas.read_csv("tall_condensed.csv", index_col = 0)
```

```python
region = tall['region']
```

```python
state, years1, years2 = 0, [], []
```

```python
#for initial testing/debugging
statements = [str(state), str(years1), str(years2)]
print 'state is ' + statements[0], 'years1 is ' + statements[1]+ ' years2 is ' + statements[2]
```

```python
#state = 0 no column names yet
#state = 1 set of possible column names found
#state = 2 column names check if paired (if match, ignore, if not, change and reset state)


for i in range(len(region)):
    try:
        if state == 1:
            if 'Table' in region.ix[i]:
                years2 = tall.ix[i+1]
                print 'years2 is ' + str(years2.to_string())
                state +=1
        if state == 2:
            if years2[1:8].all() == years1[1:8].all():
                continue
            elif years2[2] != years1[2]:
                print 'years 2 vs. years 1 is ' + str(years2), str(years1)
                chunk = tall[0:i-1] #lop off a piece
                chunk.to_csv("test1.csv")
                             
                newyears = [str(i) for i in years2]
                newyears[0] = 'region'
                tall.columns = [newyears]
                tall = tall[i+2:-1] #return what's left to continue search
                tall.to_csv("tall_chopped.csv")
                            
                years2 = []
                years1 = newyears
                state = 0 
        if state == 0:
            if 'Table' in region.ix[i]:
                years1 = tall.ix[i+1]
                state += 1
            
    except TypeError:
        continue
                
```
